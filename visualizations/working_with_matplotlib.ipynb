# Matplotlib Workflow
1. Prepare data
2. Setup Plot
3. plot data
4. Customize plots
5. Save / show plot

```python
# prepare some data
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

# set up the figure and axis
fig, ax = plt.subplots(figsize=(10, 5))

# plot data
ax.plot(x, y)

# customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# save and show the figure

plt.show()
plt.savefig('../images/sample-plot.png')
```

# Health Datasets

```python
heart_disease = pd.read_csv("../datasets/heart-disease.csv")  
heart_disease  
```

```python
heart_disease["age"].plot.hist(bins=50);
```

```python
heart_disease.plot.hist(figsize=(10, 30), subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind='scatter', x='age', y='chol', c='target');
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);

ax.set_xlim([45, 100]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol");

ax.legend(*scatter.legend_elements(), title="Target");
ax.axhline(over_50["chol"].mean(), linestyle="-");
```

```python
over_50.head()
```

```python
fig, (ax0, ax1) = plt.subplots(figsize=(10, 10), nrows=2, ncols=1, sharex=True)

ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter")
ax0.set(title="Heart Disease and Cholesterol Levels", ylabel="Cholesterol")

ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(over_50["chol"].mean(), linestyle="--");


ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="winter")
ax1.set(title="Heart Disease and Max Heart Rate", ylabel="Max Heart Rate", xlabel="Age")

ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(), linestyle="--");


fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
